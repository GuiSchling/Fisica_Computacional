```python
import numpy as np
import matplotlib.pyplot as plt

def a(x, k, m):
    w2 = k/m
    a =  -w2 * x
    return a 

def euler(f_acel, r0, v0, t_array, m, k):
    n = len(t_array)
    r = np.zeros(n) 
    v = np.zeros(n) 
    
    r[0] = r0
    v[0] = v0
    
    for i in range(n - 1):
        dt = t_array[i+1] - t_array[i]
        

        a_atual = f_acel(r[i], k, m)
        

        r[i+1] = r[i] + v[i] * dt
        v[i+1] = v[i] + a_atual * dt
        
    return r, v
```

```python
# Parâmetros
m = 1.0  # kg
k = 10.0 # N/m
r0 = 0.5 # deslocamento inicial de 50cm
v0 = 0.0 # partindo do repouso
t = np.linspace(0, 10, 1000)


# Execução
posicao, velocidade = euler(a, r0, v0, t, m, k)

# Plot
plt.plot(posicao, velocidade, label='Posição (x)')
plt.axhline(0, color='black', lw=1, ls='--')


```

```python
def energia(r, v, m, k):
    ec = 0.5 * m * v**2
    ep = 0.5 * k * r**2
    em = ec + ep
    return ec, ep, em

# Após rodar o seu euler_massa_mola:
ec, ep, em = energia(posicao, velocidade, m, k)

plt.yscale("log")
plt.plot(t, em)
```
