```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def euler(f, t0, y0,tf,dt):
    '''
    f: função taxa
    t0: tempo inicial
    y0: ponto de partida
    tf: tempo final
    dt: passo de tempo
    '''
    '''yn+1 = yn + h . f(t,y) '''
    n_passos = int((tf-t0)/dt)
    y = y0
    t = t0
    tempos = [t0]
    valores = [y0]

    for j in range(n_passos):
        y = y + dt * f(t, y)
        t += dt
        tempos.append(t)
        valores.append(y)
    return tempos, valores
        
```

```python

def aceleracao(t,v):
    k = 0.5 #coeficiente de resistência do ar
    g = 9.8 #acerelação gravitacional

    return -g - k*v
```

```python
tempos_v, velocidades = euler(aceleracao,0,0,10,0.1)

def simulacao_velocidade(t,y):
    return np.interp(t,tempos,velocidades)

tempos, posicoes = euler(simulacao_velocidade,0,100,10,0.1 )

plt.style.use('dark_background')
fig, ax =  plt.subplots(1,2,figsize=(12,5))
ax[0].plot(tempos, posicoes, color='red')
ax[0].set_ylim(0,110)
ax[0].set_xlim(0,10)

ax[1].plot(tempos_v, velocidades, color='cyan')
```
