```python
import numpy as np

def f1(x, y):  
    return x**2 + y**2 - 4

def f2(x, y): 
    return np.exp(x) + y - 1

def jacobian(x, y):
    return np.array([[2*x, 2*y], [np.exp(x), 1]])

def newton_raphson_2d(f1, f2, jacobian, x0, y0):
    x, y = x0, y0
    for _ in range(100):
        J = jacobian(x, y)
        f = np.array([f1(x, y), f2(x, y)])
        delta = np.linalg.solve(J, f)
        print(delta)
        x_new, y_new = x - delta[0], y - delta[1]
        if np.linalg.norm([x_new - x, y_new - y]) < 1e-6:
            return x_new, y_new
        x, y = x_new, y_new
    return None

print(newton_raphson_2d(f1,f2,jacobian, 5,5))
```
