```python
import numpy as np

def derivada(f,x):
    '''Função que calcula a derivada de f no ponto x usando o método de diferenças finitas'''
    h = 1e-6
    return (f(x+h)-(f(x)))/h


def newthon_raphson(f,df,x,N,eps = 1e-4):
    ''' 
    f: função a ser trabalhada
    df: função que calcula a derivada de f
    x: entrada de x inicial
    N: número de iterações
    eps: tolerância para o critério de parada
    '''

    for i in range (N):
        if abs(f(x)) < eps:
            return x, True
            
        x = x- (f(x)/df(f,x))
    return x, False


```

```python
#Funções de teste
def funcao(z):
    return z**3 +1 
```

```python
#Teste para encontrar uma raíz
z0 = complex(1,1)
print(newthon_raphson(funcao,derivada, z0, 100))
```

```python
#Para encontrar mais de uma raíz
raizes = []
z_inicial = np.arange(-2,2,1)
for r in z_inicial: #Eixo dos reais para chute inicial
    for i in z_inicial: #Eixo dos imaginários para chute inicial
        z0 = complex(r,i)
        if z0 ==0j: #Prevenir divisão por zero
            continue
        raiz, status = newthon_raphson(funcao,derivada, z0, 100)
        if status:
            raizes.append(complex(round(raiz.real,2), round(raiz.imag,2)))

raizes = list(set(raizes))
print(raizes)
```
