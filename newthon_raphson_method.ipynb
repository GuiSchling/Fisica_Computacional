```python
import numpy as np

def derivada(f,x):
    h = 1e-6
    return (f(x+h)-(f(x)))/h


def newthon_raphson(f,x,N,eps = 1e-4):
    ''' 
    f: função a ser trabalhada
    x: entrada de x inicial
    N: número de iterações
    eps: tolerância para o critério de parada
    '''

    for i in range (N):
        if abs(f(x)) < eps:
            return x, True
            
        x = x- (f(x)/derivada(f,x))
    return x, False


```

```python
# Funções de teste
def funcao(z):
    return z**3 +1 
```

```python
# Teste para encontrar uma raíz
z0 = complex(1,1)
print(newthon_raphson(funcao, z0, 100))
```

```python
raizes = []
for r in np.arange(-2,2,1):
    for i in np.arange(-2,2,1):
        z0 = complex(r,i)
        if z0 ==0j:
            continue
        raiz, status = newthon_raphson(funcao, z0, 100)
        if status:
            raizes.append(complex(round(raiz.real,2), round(raiz.imag,2)))

raizes = list(set(raizes))
print(raizes)
```
