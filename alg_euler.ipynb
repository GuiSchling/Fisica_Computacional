```python
import matplotlib.pyplot as plt
import numpy as np
```

```python

def f(x, k):
    '''Função recebe o ponto X e o valor da constante 'k' e retorna o valor da função'''
    return -k * x

```

```python
x0 = 100 # Quantidade inicial de material
k = 0.2 # Constante de decaimento
t_total = 20 # Tempo total de simulação
dt = 0.1 # Passo de tempo
n_passos = int(t_total / dt) # Número de passos de tempo

tempos = np.linspace(0, t_total, n_passos) # Vetor de tempos
quantidades = np.zeros(n_passos) # Vetor para armazenar as quantidades de material
quantidades[0] = x0 # Condição inicial
```

```python
for i in range(1, n_passos):
    quantidades[i] = quantidades[i-1] + f(quantidades[i-1], k) * dt
```

```python
import matplotlib.pyplot as plt

plt.plot(tempos, quantidades, label='Quantidade de material')
plt.xlabel('Tempo')
plt.ylabel('Quantidade')
plt.title('Decaimento Radioativo - Método de Euler')
plt.grid()
plt.show()  
```
