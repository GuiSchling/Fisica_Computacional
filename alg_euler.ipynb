```python
import matplotlib.pyplot as plt
import numpy as np
```

```python

def f(x):
    '''Função recebe o ponto X e o valor da constante 'k' e retorna o valor da função'''
    return -x

```

```python
def euler(x0, f, dt, t_total):
    '''Função recebe a quantidade inicial, a função de taxa de variação, o passo de tempo e o tempo total, e retorna um array com as quantidades ao longo do tempo'''
    n_passos = int(t_total / dt)
    quantidades = np.zeros(n_passos)
    quantidades[0] = x0
    for i in range(1, n_passos):
        quantidades[i] = quantidades[i-1] + f(quantidades[i-1]) * dt
    return quantidades
```

```python
import matplotlib.pyplot as plt

def plotar_slope_field(tempos, q_iniciais):
    for q in q_iniciais:
        plt.plot(tempos, q, label='Quantidade de material')
    plt.xlabel('Tempo')
    plt.ylabel('Quantidade')
    plt.title('Slope Field -                                                                                                                      Método de Euler')
    plt.grid()
    plt.show()  
```

```python
#configurações iniciais
dt = 0.1 # Passo de tempo
t_total = 10 # Tempo total
tempos = np.linspace(0, t_total, int(t_total / dt)) # Vetor de tempos
```

```python
quantidade_lista = []
for x0 in range(-10,10,1):
    quantidade = euler(x0, f, dt, t_total)
    quantidade_lista.append(quantidade)

plotar_slope_field(tempos, quantidade_lista)
```

```python
# Exercício 2

def y2(x):
    return -x+1


q_iniciais = []
for x0 in range(-10,10,1):
    quantidade = euler(x0, y2, dt, t_total)
    q_iniciais.append(quantidade)

plotar_slope_field(tempos, q_iniciais)
```
